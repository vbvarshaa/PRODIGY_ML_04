```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Correct Kaggle path
path = "/kaggle/input/datasets/organizations/gti-upm/leapgestrecog/leapGestRecog"

data = []
labels = []

limit = 100

for person in os.listdir(path):
    person_path = os.path.join(path, person)
    
    for gesture in os.listdir(person_path):
        gesture_path = os.path.join(person_path, gesture)
        
        count = 0
        
        for img in os.listdir(gesture_path):
            
            if count >= limit:
                break
            
            img_path = os.path.join(gesture_path, img)
            
            try:
                image = cv2.imread(img_path)
                
                if image is None:
                    continue
                
                image = cv2.resize(image, (64, 64))
                image = image.flatten()
                
                data.append(image)
                labels.append(gesture)
                
                count += 1
                
            except:
                continue

data = np.array(data)
labels = np.array(labels)

print("Data loaded:", data.shape)

# Encode labels
le = LabelEncoder()
labels = le.fit_transform(labels)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.2, random_state=42
)

# Train
model = SVC(kernel='linear')
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)

# Test prediction
pred = model.predict([X_test[0]])
print("Predicted Gesture:", le.inverse_transform(pred))
```
